# Problem 1 - Data Visualization in Gait Maturation
The data contained here are from 50 healthy children ranging in age from 40 months to 163 months. In very young children, immature control of posture and gait results in unsteady locomotion. In children of approximately 3 yr of age, gait appears relatively mature; however, it is unknown whether the dynamics of walking change beyond this age. A summary of pertinent data about the subjects is presented in the following csv file.

Reference: JM Hausdorff, L Zemany, C.-K. Peng and AL Goldberger, Maturation of Gait Dynamics: Stride-to-Stride Variability and Its Temporal Organization in Children, Journal of Applied Physiology 86:1040-1047, 1999.

**Step 1. Import the necessary libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns

# %matplotlib inline sets the backend of matplotlib to the 'inline' backend:
# With this backend, the output of plotting commands is displayed inline within
# frontends like the Jupyter notebook, directly below the code cell that produced it.
# The resulting plots will then also be stored in the notebook document.
%matplotlib inline
```

**Step 2. Import the dataset from this [address](https://physionet.org/files/gait-maturation-db/1.0.0/data/table.csv?download)**

**Step 3. Assign it to a variable called gait**

```python
gait = pd.read_csv('https://physionet.org/files/gait-maturation-db/1.0.0/data/table.csv?download')
gait
```

**Step 4. Plot counts of the age group column using Pandas**

```python
gait.Group.value_counts().plot(kind='bar')
```

**Step 5. Generate the following bar plot of median Speed-m/sec values based on age group **

```python
gait.groupby('Group')['Speed-m/sec'].agg(np.median).plot(kind = 'bar')
```

**Step 6. Box plot example**

```python
g = sns.catplot("gender", "Speed-m/sec", "Group", gait, kind="box",                        
                   palette="PRGn",aspect=2)
g.set(ylim=(0.7, 1.6))
```

```python
g = sns.catplot("height-inches", "Speed-m/sec", "Group", gait, kind="box",                        
                   palette="PRGn",aspect=3)
g.set(ylim=(0.7, 1.6))
```

**Step 7. Scatter plot example**

```python
sns.catplot("Group", "leg-length-inches", data=gait, palette="BuPu_d")
```

**Step 8. Regression plot**

```python
sns.regplot("leg-length-inches", "Speed-m/sec", data=gait, robust=True, ci=95, color="seagreen")
sns.despine();
```

```python
sns.pairplot(gait)
```
